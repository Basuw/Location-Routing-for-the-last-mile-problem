```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import time
from tqdm import tqdm
```

```python
# Define a semente para reproduzir os resultados
np.random.seed(42)

#Carrega os dados dos clientes
df = pd.read_excel('C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\dados.xlsx')

# Cria lista de datas
datas = pd.date_range(start='2020-01-01', end='2020-12-31', freq='D')
df_datas = pd.DataFrame({'Data': datas})

# Cria tabela dia a dia para todos os clientes
df_demanda = df.merge(df_datas, how='cross')

# Define o número de instâncias que serão simuladas
num_instancias = 1

# Cria um array tridimensional para armazenar as demandas diárias simuladas
demanda = np.zeros((len(df), len(datas), num_instancias))

# Simula a demanda para cada cliente, para cada dia e para cada instância
for j, data in tqdm(enumerate(datas), desc='Simulando demandas'):
    for i, cliente in enumerate(df['Order Postal Code']):
        for n in range(num_instancias):
            # Simula uma demanda diária para o cliente e a data atuais
            is_dem = stats.bernoulli.rvs(0.1)
            dem = np.random.poisson(3)

            # Armazena a demanda diária simulada na matriz tridimensional
            demanda[i, j, n] = is_dem * dem


# Calcula as médias e máximas das demandas diárias simuladas para cada cliente e data
media_demanda = np.mean(demanda, axis=2).flatten()
maior_demanda = np.max(demanda, axis=2).flatten()

# Calcula a quantidade de instâncias com demanda para cada dia de cada cliente
qtd_demanda = np.count_nonzero(demanda, axis=2)
qtd_inst_demanda = qtd_demanda.flatten()

# Cria um DataFrame para armazenar as informações resumidas das demandas
df_resumo = pd.DataFrame({
    'Order Postal Code': np.repeat(df['Order Postal Code'], len(datas)),
    'Data': np.tile(datas, len(df)),
    'media_demanda': media_demanda,
    'maior_demanda': maior_demanda,
    'qtd_inst_demanda': qtd_inst_demanda
})

df_resumo.to_csv('resumo_demandas.csv', sep=';')
pd.set_option('display.max_columns', None)
```
