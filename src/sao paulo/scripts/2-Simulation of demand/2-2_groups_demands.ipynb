```python
import pandas as pd
```

```python
# DataFrames
df_demand = pd.read_csv("../../results/utils/demand_instances.csv", usecols=['Order Postal Code', 'Date', 'instance_0', 'instance_1', 'instance_2', 'instance_3', 'instance_4'])
df_clients_grids = pd.read_csv("../../results/utils/df_clients_grids.csv")

# Merge dataframes on "Order Postal Code"
df_merged = df_demand.merge(df_clients_grids, how='left')

df = df_merged[['Order Postal Code', 'Date', 'Grid', 'instance_0', 'instance_1', 'instance_2', 'instance_3', 'instance_4']]
df = df.drop([ 'Order Postal Code'], axis=1)
df_grouped = df.groupby(["Grid", "Date"]).sum().reset_index()

df_grouped.to_csv('../../results/utils/df_square_demand.csv')
```
