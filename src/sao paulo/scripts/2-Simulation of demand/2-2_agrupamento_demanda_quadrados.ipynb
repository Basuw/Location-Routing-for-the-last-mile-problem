```python
import pandas as pd
```

```python
# DataFrames
df_demand = pd.read_csv("../../results/utils/demands.csv", usecols = ['Order Postal Code','Date', 'instancia_0', 'instancia_1', 'instancia_2', 'instancia_3', 'instancia_4', 'instancia_5'])
df_clients_grids = pd.read_csv("../../results/utils/df_clients_grids.csv")

# Merge dataframes on "Order Postal Code"
df_merged = df_demand.merge(df_clients_grids, how='left')

df = df_merged[['Order Postal Code','Date', 'Grid', 'instancia_0', 'instancia_1', 'instancia_2', 'instancia_3', 'instancia_4', 'instancia_5']]
df = df.drop([ 'Order Postal Code'], axis=1)
df_grouped = df.groupby(["Grid", "Date"]).sum().reset_index()

df_grouped.to_csv('../../results/utils/df_demanda_quadrados.csv')
```
