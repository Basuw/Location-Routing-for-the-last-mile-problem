```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import time
from tqdm import tqdm
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Load client data
df = pd.read_excel('../../data/data.xlsx')

# Create date range
dates = pd.date_range(start='2020-01-01', end='2020-12-31', freq='D')
df_dates = pd.DataFrame({'Date': dates})

# Create daily table for all clients
df_demand = df.merge(df_dates, how='cross')

# Define the number of instances to simulate
n_instances = 1

# Create 3D array to store simulated daily demands
demand = np.zeros((len(df), len(dates), n_instances))

# Simulate demand for each client, day, and instance
for j, data in tqdm(enumerate(dates), desc='Simulating demands'):
    for i, client in enumerate(df['Order Postal Code']):
        for n in range(n_instances):
            # Simulate daily demand for current client and date
            is_dem = stats.bernoulli.rvs(0.1)
            dem = np.random.poisson(3)

            # Store simulated daily demand in 3D array
            demand[i, j, n] = is_dem * dem


# Compute mean and max simulated daily demands per client and date
avg_demand = np.mean(demand, axis=2).flatten()
max_demand = np.max(demand, axis=2).flatten()

# Count instances with non-zero demand per client per day
demand_count = np.count_nonzero(demand, axis=2)
demand_instance_count = demand_count.flatten()

# Create summary DataFrame
df_summary = pd.DataFrame({
    'Order Postal Code': np.repeat(df['Order Postal Code'], len(dates)),
    'Date': np.tile(dates, len(df)),
    'avg_demand': avg_demand,
    'max_demand': max_demand,
    'demand_instance_count': demand_instance_count
})

df_summary.to_csv('../../results/utils/demands.csv', sep=';')
pd.set_option('display.max_columns', None)
```
