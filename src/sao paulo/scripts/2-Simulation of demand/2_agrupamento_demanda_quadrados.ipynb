```python
import pandas as pd
```

```python
# Dataframe
df_demanda = pd.read_csv("'C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\Resultados\\Utils\\resumo_demanda.csv", usecols = ['Order Postal Code','Data', 'instancia_0', 'instancia_1', 'instancia_2', 'instancia_3', 'instancia_4', 'instancia_5'])
df_clientes_quadrados = pd.read_csv("'C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\Resultados\\Utils\\df_clientes_quadrados.csv")

# Realizar a junção dos dataframes utilizando o indicador "Order Postal Code"
df_merged = df_demanda.merge(df_clientes_quadrados, how='left')

df = df_merged[['Order Postal Code','Data', 'Quadrado', 'instancia_0', 'instancia_1', 'instancia_2', 'instancia_3', 'instancia_4', 'instancia_5']]
df = df.drop([ 'Order Postal Code'], axis=1)
df_grouped = df.groupby(["Quadrado", "Data"]).sum().reset_index()

df_grouped.to_csv('C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\Resultados\\Utils\\df_demanda_quadrados.csv')
```
