```python
import pandas as pd
from itertools import product, repeat
from haversine import haversine, Unit
import numpy as np
import plotly.express as px
from tqdm import tqdm
```

```python
df_candidates = pd.read_excel('C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\dados.xlsx', sheet_name='candidates')
df_quadrados = pd.read_csv('C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\Resultados\\Utils\\df_quadrados.csv')

df_quadrados['Latitude'] = (df_quadrados['maxLat'] + df_quadrados['minLat'])/2
df_quadrados['Longitude'] = (df_quadrados['maxLong'] + df_quadrados['minLong'])/2
```

```python
cols = ['quadrado'] + list(df_candidates['Nome'])

df_distancias_cds = pd.DataFrame(columns=cols)

for i, row in df_quadrados.iterrows():
    quadrado = row['id']
    distancias = []
    for j, candidate_row in df_candidates.iterrows():
        candidate = candidate_row['Nome']
        coords_quadrado = (row['Latitude'], row['Longitude'])
        coords_candidate = (candidate_row['Latitude'], candidate_row['Longitude'])
        distancia_km = haversine(coords_quadrado, coords_candidate, unit=Unit.KILOMETERS)
        distancias.append(distancia_km)
    df_distancias_cds.loc[i] = [quadrado] + distancias

df_result = df_distancias_cds.melt(id_vars=["quadrado"], var_name="CD", value_name="Distance")
df_result.to_csv('C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\Resultados\\Utils\\df_distancias_cds.csv')
```

```python
cols = ['quadrado1'] + list(df_quadrados['id'])
df_distancias_quadrados = pd.DataFrame(columns=cols)

for i, quadrado1 in df_quadrados.iterrows():
    quadrado = quadrado1['id']
    distancias = []
    for j, quadrado2 in df_quadrados.iterrows():
        candidate = quadrado2['id']
        coords_quadrado = (quadrado1['Latitude'], quadrado1['Longitude'])
        coords_candidate = (quadrado2['Latitude'], quadrado2['Longitude'])
        distancia_km = haversine(coords_quadrado, coords_candidate, unit=Unit.KILOMETERS)
        distancias.append(distancia_km)
    df_distancias_quadrados.loc[i] = [quadrado] + distancias

df_result = df_distancias_quadrados.melt(id_vars=["quadrado1"], var_name="quadrado2", value_name="Distance")
df_result.to_csv('C:\\Users\\Lenovo\\OneDrive\\Documentos\\André Luis\\1. UFMG\\TCC\\TCC\\Resultados\\Utils\\df_distancias_quadrados.csv')
```
