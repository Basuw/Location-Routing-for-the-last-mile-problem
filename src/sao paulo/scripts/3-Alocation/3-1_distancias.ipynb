```python
import pandas as pd
from itertools import product, repeat
from haversine import haversine, Unit
import numpy as np
import plotly.express as px
from tqdm import tqdm
```

```python
df_candidates = pd.read_excel('../../data/dados.xlsx', sheet_name='candidates')
df_grids = pd.read_csv('../../results/utils/df_grids.csv')

df_grids['Latitude'] = (df_grids['maxLat'] + df_grids['minLat'])/2
df_grids['Longitude'] = (df_grids['maxLong'] + df_grids['minLong'])/2
```

```python
cols = ['grid'] + list(df_candidates['Nome'])

df_dist_dcs = pd.DataFrame(columns=cols)

for i, row in df_grids.iterrows():
    cell = row['id']
    distances = []
    for j, candidate_row in df_candidates.iterrows():
        candidate = candidate_row['Nome']
        grid_coords = (row['Latitude'], row['Longitude'])
        coords_candidate = (candidate_row['Latitude'], candidate_row['Longitude'])
        distance_km = haversine(grid_coords, coords_candidate, unit=Unit.KILOMETERS)
        distances.append(distance_km)
    df_dist_dcs.loc[i] = [cell] + distances

df_result = df_dist_dcs.melt(id_vars=["grid"], var_name="CD", value_name="Distance")
df_result.to_csv('../../results/utils/df_dist_dcs.csv')
```

```python
cols = ['grid1'] + list(df_grids['id'])
df_dist_grids = pd.DataFrame(columns=cols)

for i, grid_row in df_grids.iterrows():
    cell = grid_row['id']
    distances = []
    for j, other_grid_row in df_grids.iterrows():
        candidate = other_grid_row['id']
        grid_coords = (grid_row['Latitude'], grid_row['Longitude'])
        coords_candidate = (other_grid_row['Latitude'], other_grid_row['Longitude'])
        distance_km = haversine(grid_coords, coords_candidate, unit=Unit.KILOMETERS)
        distances.append(distance_km)
    df_dist_grids.loc[i] = [cell] + distances

df_result = df_dist_grids.melt(id_vars=["grid1"], var_name="grid2", value_name="Distance")
df_result.to_csv('../../results/utils/df_dist_grids.csv')
```
